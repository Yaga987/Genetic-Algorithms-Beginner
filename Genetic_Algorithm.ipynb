*Import Package*

```python
import random
```

*Problem*

```python
def foo(a,b,c):
  return (8*a**3 + 7*c**2 + 59*b**5 + 146) - 1824   
  #In our case our problem answer is 1824
```

*Fitness Method*

```python
def checker(a,b,c):
  return (8*a**3 + 7*c**2 + 59*b**5 + 146)
```

```python
def fitness(a,b,c):
  ans = foo(a,b,c)

  if ans ==0:
    return 99999
  else:
    return abs(1/ans)
```

```python
t = 10000
```

*Generate Solution*

```python
sol = []
for x in range(t):
  sol.append(
              (
              random.uniform(0,t),
              random.uniform(0,t),
              random.uniform(0,t)
              )
             )

print(sol[:5])
```

```python
for i in range(t):

  rankedsol = []

  for y in sol:
    rankedsol.append((fitness(y[0],y[1],y[2]), y))

  rankedsol.sort()
  rankedsol.reverse()

  print(f'=== Gen{i} best solution === ')
  print(rankedsol[0])

  if rankedsol[0][0] > 999:
    break

  bestsol = rankedsol[:100]

  elements = []

  for z in bestsol:
    elements.append(z[1][0])
    elements.append(z[1][1])
    elements.append(z[1][2])

  newGen = []

  # Mutate

  for f in range(1000):
    e1 = random.choice(elements) * random.uniform(0.99, 1.01)
    e2 = random.choice(elements) * random.uniform(0.99, 1.01)
    e3 = random.choice(elements) * random.uniform(0.99, 1.01)

    newGen.append((e1,e2,e3))

  sol = newGen
```
